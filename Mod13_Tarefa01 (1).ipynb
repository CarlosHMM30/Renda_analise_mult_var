```python
import pandas as pd
from patsy import dmatrices
import numpy as np
import statsmodels.api as sm
```

```python
df = pd.read_csv('previsao_de_renda.csv')
```

1. Ajuste um modelo para prever log(renda) considerando todas as covariáveis disponíveis.
    - Utilizando os recursos do Patsy, coloque as variáveis qualitativas como *dummies*.
    - Mantenha sempre a categoria mais frequente como casela de referência
    - Avalie os parâmetros e veja se parecem fazer sentido prático.  


2. Remova a variável menos significante e analise:
    - Observe os indicadores que vimos, e avalie se o modelo melhorou ou piorou na sua opinião.
    - Observe os parâmetros e veja se algum se alterou muito.  


3. Siga removendo as variáveis menos significantes, sempre que o *p-value* for menor que 5%. Compare o modelo final com o inicial. Observe os indicadores e conclua se o modelo parece melhor. 
    

```python
y, x = dmatrices('np.log(renda)~ C(sexo) + C(tipo_renda) + posse_de_veiculo + posse_de_imovel + qtd_filhos + C(educacao) + C(estado_civil ) + C(tipo_residencia) + idade + tempo_emprego + qt_pessoas_residencia + renda', data=df) 
x
```

```python
sm.OLS(y, x).fit().summary()
```

```python
y, x = dmatrices('np.log(renda)~ C(sexo, Treatment(1)) + C(tipo_renda) + posse_de_veiculo + posse_de_imovel + qtd_filhos + C(educacao, Treatment(2)) + C(estado_civil ) + C(tipo_residencia,Treatment(1)) + idade + tempo_emprego + qt_pessoas_residencia + renda', data=df) 
x
```

```python
df['sexo'].value_counts(), df['tipo_renda'].value_counts(), df['educacao'].value_counts(), df['estado_civil'].value_counts(), df['tipo_residencia'].value_counts()
```

Avaliando os parâmetros acima é possível concluir que não faz sentido utilizar todas elas no modelo, existem várias variáveis que não são significativas.

```python
sm.OLS(y, x).fit().summary()
```

A variável menos significativa é a Educação T.Superior incompleto, então a variável Educação será retirada do modelo

```python
y, x = dmatrices('np.log(renda)~ C(sexo, Treatment(1)) + C(tipo_renda) + posse_de_veiculo + posse_de_imovel + qtd_filhos + C(estado_civil ) + C(tipo_residencia,Treatment(1)) + idade + tempo_emprego + qt_pessoas_residencia + renda', data=df) 
```

```python
sm.OLS(y, x).fit().summary()
```

Houve pouca variação do R². Desta forma, não aconteceu impacto significativo no modelo.

```python
y, x = dmatrices('np.log(renda)~ C(sexo, Treatment(1))  + posse_de_veiculo + posse_de_imovel + qtd_filhos + idade + tempo_emprego + qt_pessoas_residencia + renda', data=df) 
```

```python
sm.OLS(y, x).fit().summary()
```

Retirando as variáveis menos significativas o modelo ficou melhor, pois ele ficou mais simples e não acarretou em uma variação significativa do R²
